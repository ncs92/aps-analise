```python
def gerar1000():
    arq = open('./arq-1000.txt', 'a')
    for x in range(1000):
        arq.write(str(randint(-999,999)) + ' ')
    arq.close  
    
def gerar10000():
    arq = open('./arq-10000.txt', 'a')
    for x in range(10000):
        arq.write(str(randint(-999,999)) + ' ')
    arq.close 
    
def gerar100000():
    arq = open('./arq-100000.txt', 'a')
    for x in range(100000):
        arq.write(str(randint(-999,999)) + ' ')
    arq.close 

def carregaVetor(arquivo):
    arq = open(arquivo)
    return [int(num) for num in arq.read().split(' ') if num != '']

def carregaArquivoVetor():
    global v1, v2, v3
    gerar1000()
    gerar10000()
    gerar100000()
    v1 = carregaVetor('./arq-1000.txt')
    v2 = carregaVetor('./arq-10000.txt')
    v3 = carregaVetor('./arq-100000.txt')
```

```python
n_inst_quick = 0

def bubbleSort(vetor):
    n_inst = 0
    tamanho = len(vetor)
    temp = 0
    v = vetor.copy()
    
    for i in range(0, tamanho-1):
        n_inst += 1
        for j in range((i+1), tamanho):
            n_inst += 1
            if (v[j] < v[i]):
                temp = v[i]
                v[i] = v[j]
                v[j] = temp 
    return v, n_inst

def insertionSort(vetor):
    v = vetor.copy()
    n_inst = 0
    for i in range(1, len(v)):
        n_inst += 1
        item = v[i]
        j = i
        while j > 0 and item < v[j - 1]:
            n_inst += 1
            v[j] = v[j - 1]
            j -= 1
        v[j] = item
    return v, n_inst

def selectionSort(vetor):
    n_inst = 0
    v = vetor.copy()
    tamanho = len(vetor)
    for i in range(0, tamanho-1):
        n_inst += 1
        min = i
        for j in range((i+1), tamanho):
            n_inst += 1
            if(v[j] < v[min]):
                min = j
        if (v[i] != v[min]):
            aux = v[i]
            v[i] = v[min]
            v[min] = aux
    return v, n_inst

def particiona(vetor, inicio, fim, pivo):
    global n_inst_quick
    vetor[pivo], vetor[fim] = vetor[fim], vetor[pivo]
    pos = inicio
    for i in range(inicio, fim):
        n_inst_quick += 1
        if vetor[i] < vetor[fim]:
            vetor[i], vetor[pos] = vetor[pos], vetor[i]
            pos += 1
    vetor[pos], vetor[fim] = vetor[fim], vetor[pos]
    return pos


def quickSort(vetor, inicio, fim):
    global n_inst_quick
    n_inst_quick += 1
    if inicio >= fim:
        return vetor
    pivo = randrange(inicio, fim + 1)
    novo_pivo = particiona(vetor, inicio, fim, pivo)
    quickSort(vetor, inicio, novo_pivo - 1)
    quickSort(vetor, novo_pivo + 1, fim)


def quick(vetor):
    v = vetor.copy()
    quickSort(v, 0, len(v) - 1)
    return vetor


def buscaBinaria(vetor, num):
    n_inst = 0
    esq = 0
    dire = len(vetor) - 1 
    v = vetor.copy()
    while(esq <= dire):
        n_inst += 1
        meio = int((esq + dire)/2)
        if (num == v[meio]):
            return meio, n_inst
        if (num < v[meio]):
            dire = meio-1
        else:
            esq = meio+1
    return -1, n_inst


def merge(esq, dire, n_inst):
    n_inst += 1
    aux = []
    while len(esq) != 0 and len(dire) != 0:
        n_inst += 1
        if esq[0] < dire[0]:
            aux.append(esq[0])
            esq.remove(esq[0])
        else:
            aux.append(dire[0])
            dire.remove(dire[0])
    if len(esq) == 0:
        aux += dire
    else:
        aux += esq
    return aux


def mergeSort(vetor, n_inst):
    n_inst += 1
    v = vetor.copy()
    tamanho = len(v)
    if tamanho == 0 or tamanho == 1:
        return v
    else:
        meio = int(tamanho / 2)
        esq = mergeSort(v[:meio])
        dire = mergeSort(v[meio:])
    return merge(esq, dire, n_inst)


if __name__ == '__main__':
    #carregaArquivoVetor()
    global n_inst_quick
    v = [1, 5, 9, 7, 8, 6]
    r = quick(v)
    print(r)
    print(n_inst_quick)
    #print(r)

    
    
```
