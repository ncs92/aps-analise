```python
from time import time

n_inst_quick = 0
n_inst_merge = 0
n_inst_maxVetor = 0
n_inst_heap = 0


def bubbleSort(v):
    n_inst = 0
    tam = len(v)
    temp = 0
    vetor = v.copy()
    for i in range(tam-1, 0, -1):
        n_inst +=1
        for j in range(i):
            n_inst += 1
            if vetor[j] > vetor[j+1]:
                vetor[j], vetor[j+1] = vetor[j+1], vetor[j]
    return vetor, n_inst

def bubbleSortMelhorado(v):
    n_inst = 0
    tam = len(v)
    temp = 0
    vetor = v.copy()
    for i in range(tam-1, 0, -1):
        trocou = False
        n_inst +=1
        for j in range(i):
            n_inst += 1
            if vetor[j] > vetor[j+1]:
                vetor[j], vetor[j+1] = vetor[j+1], vetor[j]
                trocou = True
        if not trocou:
            return vetor, n_inst


def insertionSort(vetor):
    v = vetor.copy()
    n_inst = 0
    for i in range(1, len(v)):
        n_inst += 1
        item = v[i]
        j = i
        while j > 0 and item < v[j - 1]:
            n_inst += 1
            v[j] = v[j - 1]
            j -= 1
        v[j] = item
    return v, n_inst

def selectionSort(vetor):
    n_inst = 0
    v = vetor.copy()
    tamanho = len(vetor)
    for i in range(0, tamanho-1):
        n_inst += 1
        min = i
        for j in range((i+1), tamanho):
            n_inst += 1
            if(v[j] < v[min]):
                min = j
        if (v[i] != v[min]):
            aux = v[i]
            v[i] = v[min]
            v[min] = aux
    return v, n_inst

def particiona(vetor, inicio, fim, pivo):
    global n_inst_quick
    vetor[pivo], vetor[fim] = vetor[fim], vetor[pivo]
    pos = inicio
    for i in range(inicio, fim):
        n_inst_quick += 1
        if vetor[i] < vetor[fim]:
            vetor[i], vetor[pos] = vetor[pos], vetor[i]
            pos += 1
    vetor[pos], vetor[fim] = vetor[fim], vetor[pos]
    return pos


def quickSort(vetor, inicio, fim):
    global n_inst_quick
    n_inst_quick += 1
    if inicio >= fim:
        return vetor
    pivo = randrange(inicio, fim + 1)
    novo_pivo = particiona(vetor, inicio, fim, pivo)
    quickSort(vetor, inicio, novo_pivo - 1)
    quickSort(vetor, novo_pivo + 1, fim)


def quick(vetor):
    v = vetor.copy()
    quickSort(v, 0, len(v))
    return vetor


def buscaBinaria(vetor, num):
    n_inst = 0
    esq = 0
    dire = len(vetor) - 1 
    v = vetor.copy()
    while(esq <= dire):
        n_inst += 1
        meio = int((esq + dire)/2)
        if (num == v[meio]):
            return meio, n_inst
        if (num < v[meio]):
            dire = meio-1
        else:
            esq = meio+1
    return -1, n_inst


def merge(esq, dire):
    global n_inst_merge
    n_inst_merge += 1
    aux = []
    while len(esq) != 0 and len(dire) != 0:
        n_inst_merge += 1
        if esq[0] < dire[0]:
            aux.append(esq[0])
            esq.remove(esq[0])
        else:
            aux.append(dire[0])
            dire.remove(dire[0])
    if len(esq) == 0:
        aux += dire
    else:
        aux += esq
    return aux


def mergeSort(vetor):
    global n_inst_merge
    n_inst_merge += 1
    v = vetor.copy()
    tamanho = len(v)
    if tamanho == 0 or tamanho == 1:
        return v
    else:
        meio = int(tamanho / 2)
        esq = mergeSort(v[:meio])
        dire = mergeSort(v[meio:])
    return merge(esq, dire)

def findMaxCrossingSubArray(a, e, m, d):
    global n_inst_maxVetor
    leftSum = float("-inf") 
    rightSum = float("-inf") 
    sum = 0
    maxLeft = 0
    maxRight = 0
    for i in range(m, e, -1):
        n_inst_maxVetor += 1
        sum = sum + a[i]
        if sum > leftSum:
            leftSum = sum
            maxLeft = i
    sum = 0
    for j in range(m+1, d):
        n_inst_maxVetor += 1
        sum = sum + a[j]
        if sum > rightSum:
            rightSum = sum
            maxRight = j
    return maxLeft, maxRight, leftSum+rightSum

def findMaximumSubArray(a, e, d):
    global n_inst_maxVetor
    n_inst_maxVetor += 1
    if e == d :
        return e, d, a[e]
    else:
        m = math.floor((e+d)/2)
        (left_e, left_d, left_sum) = findMaximumSubArray(a, e, m)
        (right_e, right_d, right_sum) = findMaximumSubArray(a, m+1, d)
        (cross_e, cross_d, cross_sum) = findMaxCrossingSubArray(a, e, m, d)
        if left_sum >= right_sum and left_sum >= cross_sum:
            return left_e, left_d, left_sum
        elif right_sum >= left_sum and right_sum >= cross_sum:
            return right_e, right_d, right_sum
        else:
            return cross_e, cross_d, cross_sum
        
def heapify(vetor, tam, i): 
    global n_inst_heap
    n_inst_heap +=1
    m = i
    e = 2 * i + 1
    d = 2 * i + 2
    if e < tam and vetor[i] < vetor[e]: 
        m = e
    if d < tam and vetor[m] < vetor[d]: 
        m = d 
    if m != i: 
        vetor[i],vetor[m] = vetor[m],vetor[i]
        heapify(vetor, tam, m) 

def heapSort(vetor): 
    global n_inst_heap
    tam = len(vetor) 
    for i in range(tam, -1, -1): 
        n_inst_heap += 1
        heapify(vetor, tam, i) 
    for i in range(tam-1, 0, -1): 
        n_inst_heap += 1
        vetor[i], vetor[0] = vetor[0], vetor[i]
        heapify(vetor, i, 0) 
        
def partition(vetor,p,r):
    global n_inst_quick
    i = (p-1)
    pivo = vetor[r]

    for j in range(p , r):
        n_inst_quick += 1
        if vetor[j] <= pivo:
            i = i+1
            vetor[i],vetor[j] = vetor[j],vetor[i]

    vetor[i+1],vetor[r] = vetor[r],vetor[i+1]
    return ( i+1 )

def quick(vetor,p,r):
    global n_inst_quick
    n_inst_quick += 1
    if p < r:
        q = partition(vetor,p,r)
        quick(vetor, p, q-1)
        quick(vetor, q+1, r)
        
        
def ordenaBubble():
    tempo_inicial_v1 = time()
    v, n = bubbleSort(v1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_bubble.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v, n = bubbleSort(v2)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_bubble.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v, n = bubbleSort(v3)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_bubble.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    
def ordenaInsertion():
    tempo_inicial_v1 = time()
    v, n = insertionSort(v1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_insertionSort.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v, n = insertionSort(v2)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_insertionSort.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v, n = insertionSort(v3)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_insertionSort.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
def ordenaSelection():
    tempo_inicial_v1 = time()
    v, n = selectionSort(v1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_selectionSort.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v, n = selectionSort(v2)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_selectionSort.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v, n = selectionSort(v3)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_selectionSort.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()    

    
def ordenaQuick():
    tempo_inicial_v1 = time()
    v11 = v1.copy()
    quick(v11, 0, len(v1)-1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_quick.txt", 'w+')
    cab = "instructions: " + str(n_inst_quick) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v11)):
        texto += str(v11[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v22 = v2.copy()
    quick(v22, 0, len(v2)-1)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_quick.txt", 'w+')
    cab = "instructions: " + str(n_inst_quick) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v22)):
        texto += str(v22[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v33 = v3.copy()
    quick(v33, 0, len(v3)-1)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_quick.txt", 'w+')
    cab = "instructions: " + str(n_inst_quick) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v33)):
        texto += str(v33[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    
def ordenaMerge():
    tempo_inicial_v1 = time()
    v = mergeSort(v1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_mergeSort.txt", 'w+')
    cab = "instructions: " + str(n_inst_merge) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v = mergeSort(v2)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_mergeSort.txt", 'w+')
    cab = "instructions: " + str(n_inst_merge) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v = mergeSort(v3)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_mergeSort.txt", 'w+')
    cab = "instructions: " + str(n_inst_merge) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
def ordenaHeap():
    tempo_inicial_v1 = time()
    v11 = v1.copy()
    heapSort(v11)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_heapSort.txt", 'w+')
    cab = "instructions: " + str(n_inst_heap) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v11)):
        texto += str(v11[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v22 = v2.copy()
    heapSort(v22)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_heapSort.txt", 'w+')
    cab = "instructions: " + str(n_inst_heap) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v22)):
        texto += str(v22[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v33 = v3.copy()
    heapSort(v33)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_heapSort.txt", 'w+')
    cab = "instructions: " + str(n_inst_heap) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v33)):
        texto += str(v33[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
def ordenaBubbleSortMelhorado():
    tempo_inicial_v1 = time()
    v, n = bubbleSortMelhorado(v1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_bubbleSortMelhorado.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v1) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v, n = bubbleSortMelhorado(v2)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_bubbleSortMelhorado.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v2) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v, n = bubbleSortMelhorado(v3)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_bubbleSortMelhorado.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v3) + "\n"
    texto = ''
    for i in range(len(v)):
        texto += str(v[i]) + ' '
    arquivo.writelines(cab + texto)
    arquivo.close()
    
def calculaPesquisaBinaria():
    tempo_inicial_v1 = time()
    pos, n = buscaBinaria(v1, 860)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_buscaBinaria.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v1) + ", pos: " + str(pos) + "\n"
    arquivo.writelines(cab)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    pos, n = buscaBinaria(v2, 860)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_buscaBinaria.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v2) + ", pos: " + str(pos) + "\n"
    arquivo.writelines(cab)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    pos, n = buscaBinaria(v3, 860)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_buscaBinaria.txt", 'w+')
    cab = "instructions: " + str(n) + ", tempo: " + str(tempo_final_v3) + ", pos: " + str(pos) + "\n"
    arquivo.writelines(cab)
    arquivo.close()
    
def calculaSubvetorMaximo():
    tempo_inicial_v1 = time()
    v11 = v1.copy()
    soma = findMaximumSubArray(v11, 0, len(v11)-1)
    tempo_final_v1 = time() - tempo_inicial_v1
    arquivo = open("resultado_1000_subVetorMaximo.txt", 'w+')
    cab = "instructions: " + str(n_inst_maxVetor) + ", tempo: " + str(tempo_final_v1) + ", soma: " + str(soma) + "\n"
    arquivo.writelines(cab)
    arquivo.close()
    
    tempo_inicial_v2 = time()
    v22 = v2.copy()
    soma = findMaximumSubArray(v22, 0, len(v22)-1)
    tempo_final_v2 = time() - tempo_inicial_v2
    arquivo = open("resultado_10000_subVetorMaximo.txt", 'w+')
    cab = "instructions: " + str(n_inst_maxVetor) + ", tempo: " + str(tempo_final_v2) + ", soma: " + str(soma) + "\n"
    arquivo.writelines(cab)
    arquivo.close()
    
    tempo_inicial_v3 = time()
    v33 = v3.copy()
    soma = findMaximumSubArray(v33, 0, len(v33)-1)
    tempo_final_v3 = time() - tempo_inicial_v3
    arquivo = open("resultado_100000_subVetorMaximo.txt", 'w+')
    cab = "instructions: " + str(n_inst_maxVetor) + ", tempo: " + str(tempo_final_v3) + ", soma: " + str(soma) + "\n"
    arquivo.writelines(cab)
    arquivo.close()
    
if __name__ == '__main__':
    #carregaArquivoVetor()
    #ordenaBubble()
    #ordenaInsertion()
    #ordenaSelection()
    #ordenaQuick()
    #ordenaMerge()
    #ordenaHeap()
    #ordenaBubbleSortMelhorado()
    #calculaPesquisaBinaria()
    calculaSubvetorMaximo()
    
```
